```python
def clean_gross_income(value):
    if pd.isna(value) or value in ['', '0', 0]:
        return 0

    # Convert to string if not already
    value = str(value)

    if re.fullmatch(r'[\d,]+', value):
        return int(float(value.replace(',', '')))

    match = re.match(r'^\$?(\d+\.?\d*)M$', value, flags=re.IGNORECASE)
    if match:
        return int(float(match.group(1)) * 1_000_000)

    match = re.match(r'^\$?(\d+\.?\d*)K$', value, flags=re.IGNORECASE)
    if match:
        return int(float(match.group(1)) * 1_000)

    match = re.search(r'(\d[\d,]*\.?\d*)', value)
    if match:
        return int(float(match.group(1).replace(',', '')))

    return 0

df['gross_income'] = df['gross_income'].apply(clean_gross_income)
df
```

```python
def clean_duration(duration):
    if pd.isna(duration) or duration in ['', '0']:
        return 0

    duration = str(duration).lower().strip()

    min_match = re.match(r'^(\d+)\s*min(?:ute)?s?$', duration)
    if min_match:
        return int(min_match.group(1))

    hm_match = re.match(r'^(\d+)\s*h(?:ou)?r?s?\s*(\d+)?\s*m(?:in)?(?:ute)?s?$', duration)
    if hm_match:
        hours = int(hm_match.group(1))
        mins = int(hm_match.group(2)) if hm_match.group(2) else 0
        return hours * 60 + mins

    dec_match = re.match(r'^(\d+\.?\d*)\s*h(?:ou)?r?s?$', duration)
    if dec_match:
        return round(float(dec_match.group(1)) * 60)

    hour_match = re.match(r'^(\d+)\s*h(?:ou)?r?s?$', duration)
    if hour_match:
        return int(hour_match.group(1)) * 60

    num_match = re.match(r'^(\d+)$', duration)
    if num_match:
        return int(num_match.group(1))

    return 0

df['duration'] = df['duration'].apply(clean_duration)

df
```

```python
# TF-IDF Vectorization
df['directors_id'] = df['directors_id'].str.extract(r'(nm\d{7})')
df = df.dropna(subset=['directors_id'])

df['content'] = (
    df['name'] + ' ' +
    (df['genre'] + ' ') * 5 +  # Boost genre importance
    df['directors_name'] + ' ' +
    df['description']*2  # Boost description importance
    + ' ' + df['stars_name'] + ' ' + df['year']
)

tfidf = TfidfVectorizer(stop_words='english', max_features=50000)
tfidf_matrix = tfidf.fit_transform(df['content'])

# Convert to dense and normalize (for cosine similarity)
dense_matrix = tfidf_matrix.toarray().astype('float32')
faiss.normalize_L2(dense_matrix)

index = faiss.IndexFlatIP(dense_matrix.shape[1])  # Inner product = cosine sim if normalized
index.add(dense_matrix)

def get_recommendations(title, top_n=5):
    try:
        # Search for movies that match the title (case insensitive)
        matches = df[df['name'].str.lower().str.contains(title.lower())]

        # If no exact match found, suggest the closest matches based on spelling
        if len(matches) == 0:
            print(f"\nNo movie/series found with '{title}'.")
            similar = get_close_matches(title, df['name'].str.lower(), n=3)
            if similar:
                # Use the first suggested movie from the list of close matches
                title = similar[0]
                print(f"\nUsing the closest match: {title.title()}")

                # Now search again using the corrected title
                matches = df[df['name'].str.lower().str.contains(title.lower())]
            else:
                print("No similar titles found.")
                return None

        # Get the index of the matched movie
        idx = matches.index[0]
        query_vector = dense_matrix[idx].reshape(1, -1)
        _, indices = index.search(query_vector, top_n + 1)

        # Get recommended movies
        rec_indices = indices[0][1:]
        recommendations = df.iloc[rec_indices][['name', 'genre', 'year', 'rating']].copy()

        # Get similarity scores for recommendations
        _, similarities = index.search(query_vector, top_n + 1)
        recommendations['similarity'] = similarities[0][1:]

        return recommendations

    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
print("="*50)
print("🎬 Welcome to the Movie & Series Recommender 🎬")
print("="*50)
user_input = input("🔍 Enter the name of a movie or series: ").strip()

if user_input:
    recommendations = get_recommendations(user_input)

    if recommendations is not None and not recommendations.empty:
        print("\n✨ Top 5 Similar Titles ✨\n")
        print(recommendations.to_string(index=False))
    else:
        print("\n🚫 Sorry, no recommendations could be made.")
else:
    print("\n⚠️ No input provided. Please enter a movie or series title.")
```
